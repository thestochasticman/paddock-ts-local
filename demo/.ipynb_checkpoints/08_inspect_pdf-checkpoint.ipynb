# Inspect the PDF report

`PaddockTS.Plotting.make_pdf` stitches every plot the pipeline
produced — topography, SILO + OzWALD climate panels, per-paddock
calendar pages, per-paddock phenology curves — into a single
A4-landscape PDF, with section title pages, PDF metadata, and vector
calendar text. This notebook locates the report and previews it
inline.

> **Prerequisite.** A pipeline run for the query below. The PDF is
> regenerated each time `get_outputs(query)` (or `make_pdf(query)`)
> runs, so it always reflects the latest plots on disk.


```python
from datetime import date
from pathlib import Path
from PaddockTS.query import Query

query = Query(
    bbox=[148.36265, -33.52606, 148.38265, -33.50606],
    start=date(2024, 1, 1),
    end=date(2024, 12, 31),
    stub="stages_demo",
)
pdf = Path(query.out_dir) / f"{query.stub}_report.pdf"
print("PDF path :", pdf)
print("  exists :", pdf.exists())
if pdf.exists():
    print(f"  size   : {pdf.stat().st_size / 1e6:.2f} MB")
```

## 1. PDF metadata

`make_pdf` writes a proper PDF info dict (Title / Author / Subject /
Keywords / Creator) so the report shows up cleanly in viewer tabs and
file-manager properties. Inspect it without a third-party PDF
library by grepping the raw bytes:


```python
import re

if pdf.exists():
    raw = pdf.read_bytes()
    for key in ("Title", "Author", "Subject", "Keywords", "Creator", "Producer"):
        m = re.search(rb'/' + key.encode() + rb'\s*\(([^)]{1,300})\)', raw)
        if m:
            val = m.group(1).decode(errors='replace')
            # Strip the UTF-16BE BOM matplotlib adds for non-ASCII strings
            if val.startswith('þÿ') or val.startswith('\xfe\xff'):
                val = val.encode('latin-1', errors='replace').decode('utf-16-be', errors='replace')
            print(f"  {key:9s}: {val}")
```

## 2. Inline preview (in-browser PDF viewer)

`IPython.display.IFrame` embeds the PDF in an `<iframe>` so the
browser's built-in PDF viewer renders it. Works in JupyterLab, the
classic notebook, and the VS Code notebook view.

> The embed below references the file by path, not by content — if
> you move or delete the file, the iframe goes blank. For a fully
> self-contained notebook, see the page-by-page render in section 3.


```python
from IPython.display import IFrame, display

if pdf.exists():
    display(IFrame(str(pdf), width=1000, height=750))
else:
    print("PDF not found — run get_outputs(query) or make_pdf(query) first.")
```

## 3. Render every page as a PNG (page-by-page preview)

Useful when the iframe doesn't work in your environment, or when you
want each page as a standalone image you can scroll through. Requires
`pdf2image` (`pip install pdf2image`) and the `poppler` binary
(`brew install poppler` on macOS, `apt install poppler-utils` on
Debian / Ubuntu).


```python
from IPython.display import display

try:
    from pdf2image import convert_from_path
    if pdf.exists():
        pages = convert_from_path(str(pdf), dpi=120)
        print(f"{len(pages)} pages — rendering each below.\n")
        for i, page in enumerate(pages, 1):
            print(f"\n── Page {i} / {len(pages)} ──")
            display(page)
    else:
        print("PDF not found.")
except ImportError:
    print("pdf2image not installed. Install with:")
    print("   pip install pdf2image")
    print("and ensure `poppler` is available (brew install poppler / apt install poppler-utils).")
```

## 4. Section structure

The report follows the section ordering defined in
[`PaddockTS.Plotting.make_pdf.SECTIONS`](https://thestochasticman.github.io/paddock-ts-local/api/plotting/#pdf-report):

1. **Cover page** — query stub, dates, bbox
2. **Landscape** — topography (elevation / flow accumulation / aspect / slope)
3. **Climate – SILO** — temperature, rainfall, radiation, ET, humidity (skipped if no SILO email)
4. **Climate – OzWALD** — temperature, precipitation, wind, radiation
5. **Satellite Calendar (SAM)** — per-year per-paddock thumbnail grid (vector text)
6. **Satellite Calendar (User)** — same, for user paddocks (skipped if no paddocks file)
7. **Phenology (SAM)** — NDVI curve per paddock with SoS / PoS / EoS markers
8. **Phenology (User)** — same, for user paddocks (skipped if no paddocks file)

Sections with no source PNGs on disk are silently dropped.
